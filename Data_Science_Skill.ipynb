## DataFrame 합치기

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index = [0,1,2,3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index = [4,5,6,7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index = [8,9,10,11])
```

## concat을 이용하여 data합치기
(DataFrame 열방향으로 합치기)

아무 옵션 없이 사용하면 열방향으로 병합

```python
result = pd.concat([df1,df2,df3])
result
```

## Key를 지정하여 병합

key를 지정하여 병합하면 다중 Index가 도어 key를 형성함

```python
df4=pd.DataFrame({'B': ['B2','B3','B6','B7'],
                  'D' : ['D2','D3','D6','D7'],
                  'F' : ['F2','F3','F6','F7']},
                index =[2,3,6,7])
```

```python
df4
```

## index를 주어 합치기
axis=1로 주면 column방향으로 합치며, index를 기준으로 합치게 된다.

```python
result = pd.concat([df1,df4], axis=1)
result
```

join=inner 옵션을 주면 index가 공통된 data만 합치고 나머지는 버리는 명령어임

```python
result = pd.concat([df1,df4], axis=1, join='inner')
result
```

df1의 index를 기준으로 공통된 data만 합치고 나머지 df4의 data는 버리는 명령어임

```python
result = pd.concat([df1,df4], axis=1, join_axes=[df1.index])
result
```

concat의 옵션중 열방향으로 합치면서 ignore_index=True라고 옵션을 잡으면 열방향으로 합친 후 기존 인덱스를 무시하고 처음부터 다시 index를 부여함

```python
left = pd.DataFrame({'key': ['K0','K4','K2','K3'],
                      'A':  ['A0','A1','A2','A3'],
                      'B':  ['B0','B1','B2','B3']})
right = pd.DataFrame({'key': ['K0','K1','K2','K3'],
                      'C':  ['C0','C1','C2','C3'],
                      'D':  ['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

## merge명령

key를 기준으로 합치기

```python
pd.merge(left,right,on='key')
```

두 데이터를 left 및 right 기준으로 데이터를 합치기

```python
pd.merge(left,right,how='left', on='key')
```

```python
pd.merge(left,right,how='right', on='key')
```

left와 right를 key를 기준으로 모두 합치고 싶을때 (합집합 기능)

```python
pd.merge(left,right,how='outer',on='key')
```

left와 right를 key를 기준으로 공통된 부분만 남기고 모두 없애고 싶을때 (교집합)

```python
pd.merge(left,right,how='inner',on='key')
```

## Pandas의 Pivot_table 학습하기

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("../data/02. sales-funnel.xlsx")
df.head()
```

```python
pd.pivot_table(df,index=['Name'])
```

Pivot Table을 이용하면 Name Column이 index가 되고 특별히 지정하지 않았다면 숫자형 데이터 칼럼만 남게되며, 이중 중복된 Name항목은 하나로 합쳐지고 Value등은 평균을 갖게 됩니다.

Index를 여러개 지정할때의 예입니다.

```python
pd.pivot_table(df,index=['Name','Rep','Manager'])
```

특정 Value만 지정해서 나타나도록 할 수도 있습니다.

```python
pd.pivot_table(df, index=['Manager','Rep'], values=['Price'])
```

values를 pivot_table로 합친경우 평균치가 기본입니다. 여기에 합계를 사용하려면 aggfunc 옵션을 사용해서 np.sum을 사용하면 됩니다.

```python
pd.pivot_table(df, index=['Manager','Rep'], values=['Price'], aggfunc=np.sum)
```

```python
pd.pivot_table(df, index=['Manager','Rep','Product'], values=['Price','Quantity'], aggfunc=[np.sum,np.mean],fill_value=0,margins=True)
```
